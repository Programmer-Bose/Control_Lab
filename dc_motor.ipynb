```python
import matplotlib.pyplot as plt
import numpy as np

#Ea(applied armature voltage)
Ea_volt = np.array([1.37, 3.37, 5.38, 7.37, 9.38, 11.37, 13.37, 13.98])

#N(rpm)
N_rpm = np.array([498, 1368, 2240, 3093, 3928, 4719, 5490, 5733])

# Plotting with markers
plt.plot(Ea_volt, N_rpm, marker='o', label='N(rpm)')


plt.grid(True)
# Adding labels and title
plt.xlabel('Ea(applied armature voltage) in Volts')
plt.ylabel('N(rpm)')
plt.title('Motor Charecteristics')

plt.ylim(0, 8000)

# Adding a legend
plt.legend(loc='upper right')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#N(rpm)
N_rpm = np.array([498, 1368, 2240, 3093, 3928, 4719, 5490, 5733])

#Eg (generator voltage)
Eg_volt = np.array([0.93, 2.56, 4.2, 5.8, 7.36, 8.83, 10.24, 10.69])

# Plotting with markers
plt.plot(N_rpm, Eg_volt, marker='o', label='Eg(generator voltage)')

plt.grid(True)
# Adding labels and title
plt.xlabel('N(rpm)')
plt.ylabel('Eg (generator voltage) in Volts')
plt.title('Generator Charecteristics')

plt.ylim(0, 12)

# Adding a legend
plt.legend(loc='upper right')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#N(rpm)
N_rpm = np.array([2502,2434,2369,2310,2248,2176])

#Eg (generator voltage)
Tm_NM = np.array([0.00186, 0.00252, 0.00317, 0.0038, 0.00444, 0.00522])

Tm_NM = Tm_NM*1000

# Plotting with markers
plt.plot(N_rpm, Tm_NM, marker='o', label='Tm_NM(Torque)')

plt.grid(True)
# Adding labels and title
plt.xlabel('N(rpm)')
plt.ylabel('Torque(Nm)*1000')
plt.title('Torque Speed Characteristics')

#plt.ylim(0, 12)

# Adding a legend
plt.legend(loc='upper right')

# Display the plot
plt.show()
```
